###Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

````python
import matplotlib.pyplot as plt
labels = 'com_droga', 'com_controle'
sizes = [dados['tratamento'].value_counts()]
explode = (0, 0.1)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
````

````python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(1, figsize=(9, 5))

values = []

for idx,name in enumerate(dados['tratamento'].value_counts().index.tolist()):
    values.append(dados['tratamento'].value_counts()[idx])

ratios = values
labels = ['com_droga', 'com_controle']
explode = [0.1, 0]

ax1.pie(ratios, autopct='%1.1f%%', labels=labels, explode=explode)

plt.show()
````
